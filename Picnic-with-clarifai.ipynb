```python
from clarifai.rest import ClarifaiApp

app = ClarifaiApp(api_key=CLARIFAI_KEY)
ingredients_model = app.models.get('food-items-v1.0')

def ingredients_on_picture(img_url):
    classification = ingredients_model.predict_by_url(img_url, min_value=0.50, max_concepts=15).get('outputs')[0].get(
        "data").get("concepts")
    used_ingredients_list = [x.get('name') for x in classification]

    return used_ingredients_list
```

```python
clarifai_labels = []

for item in ingredients_model.get_info(verbose=True)["model"]["output_info"]["data"]["concepts"]:
    clarifai_labels.append(item["name"])

with open('clarifai-labels.tsv', mode='w') as label_file:
    csv_writer = csv.writer(label_file, delimiter='\t')
    for item in clarifai_labels:
        print(item)
        csv_writer.writerow([item])
```

```python
from cloudinary.uploader import upload

import os

test_images = os.listdir('input/test')

online_images = dict()

for filename in test_images:
    with open("input/test/" + filename, "rb") as image:
        image = image.read()
    upload_result = upload(image, upload_preset=UPLOAD_PRESET, api_key=API_KEY,
                       api_secret=API_SECRET,
                       cloud_name=CLOUD_NAME)
    img_url = upload_result['secure_url']
    online_images[filename] = img_url
    print(img_url)
    
with open('online-images.tsv', mode='w') as images_file:
    csv_writer = csv.writer(images_file, delimiter='\t')
    for key, value in online_images.items():
        print(key + ", " + value)
        csv_writer.writerow([key, value])
```

```python
def most_common(lst):
    return max(set(lst), key=lst.count)
```
